```python

# Step 2: Preprocess the text in the 'tags' column (convert to lowercase, remove punctuation, and tokenize)
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove non-alphabet characters and punctuation
    text = re.sub(r'\W+', ' ', text)
    
    # Tokenize and remove stopwords
    tokens = [word for word in text.split() if word not in stopwords]
    
    return tokens

# Apply the preprocessing to the 'tags' column
new_df['cleaned_tags'] = new_df['tags'].apply(preprocess_text)
```

```python
# Step 3: Count the frequency of each word across all movies
all_words = []
for tags in new_df['cleaned_tags']:
    all_words.extend(tags)

# Count the frequency of each word using Counter
word_count = Counter(all_words)
```

```python

# Step 4: Select the top 5000 most frequent words
top_5000_words = [word for word, count in word_count.most_common(5000)]
```

```python

# Step 5: Create a manual binary vector for each movie
def vectorize(tags, vocabulary):
    vector = [1 if word in tags else 0 for word in vocabulary]
    return vector

# Apply the vectorization to the 'cleaned_tags' column
new_df['vector'] = new_df['cleaned_tags'].apply(lambda tags: vectorize(tags, top_5000_words))

# Convert the list of vectors into a NumPy array (to mimic the output of cv.fit_transform().toarray())
vectors = np.array(new_df['vector'].tolist())
```

```python
# Function to compute cosine similarity between two vectors
def cosine_similarity_manual(A, B):
    dot_product = np.dot(A, B)
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)
    
    return dot_product / (norm_A * norm_B)

# Compute pairwise cosine similarity for all vectors
def compute_cosine_similarity_matrix(vectors):
    n = len(vectors)
    similarity_matrix = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            similarity_matrix[i][j] = cosine_similarity_manual(vectors[i], vectors[j])
    
    return similarity_matrix

# Calculate the cosine similarity matrix for the movie vectors
similarity = compute_cosine_similarity_matrix(vectors)

# Check the shape of the similarity matrix
similarity[1]
```

```python
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Assuming your 'movies' dataframe is already loaded and preprocessed
# We'll focus on the 'overview' feature for now, but you can extend this to other features like genres or keywords

# 1. Fill missing overviews with an empty string
movies['overview'] = movies['overview'].fillna('')

# 2. Vectorize the overview text using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')  # Remove common English stop words
tfidf_matrix = tfidf.fit_transform(movies['overview'])  # Learn the vocabulary and return term-document matrix

# 3. Compute cosine similarity matrix between all movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# 4. Build a function that takes a movie title as input and returns the top 10 most similar movies
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = movies[movies['title'] == title].index[0]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the 10 most similar movies
    sim_scores = sim_scores[1:11]  # Exclude the first one (itself)
    movie_indices = [i[0] for i in sim_scores]

    # Return the titles of the top 10 most similar movies
    return movies['title'].iloc[movie_indices]

# Example: Get recommendations for 'Avatar'
recommendations = get_recommendations('Avatar')
print(recommendations)
```
